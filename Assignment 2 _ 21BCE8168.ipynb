* COUNT PLOT

```python
countplot=sns.relplot(data=dataset,x="speeding",y="total", hue="abbrev")
countplot
plt.figure(figsize=(8, 6))
```

Inference: The relational plot illustrates the relationship between speeding-related car crashes and total car crashes, with each point representing a state and color-coded by its abbreviation. It provides an overview of how speeding correlates with the total number of crashes in different states, allowing for easy comparison.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='alcohol', y='not_distracted', data=dataset)
plt.title('Car Crashes vs. Percentage of Non-Distracted Drivers')
plt.xlabel('Alcohol')
plt.ylabel('Not Distracted')
plt.show()
```

The scatter plot comparing alcohol involvement with the percentage of non-distracted drivers in car crashes suggests that there is a general negative correlation between these two variables. In other words, as alcohol involvement in car crashes increases, the percentage of non-distracted drivers tends to decrease, indicating that alcohol-related accidents are often associated with higher levels of driver distraction.

```python
sns.scatterplot(x="ins_premium", y="no_previous",palette="ch:r=-.2,d=.3_r",sizes=(1, 8), linewidth=0, data=dataset )
```

The plot suggests that states with lower insurance premium costs tend to have a higher prevalence of accidents where drivers have no previous accidents on their records. This could indicate that lower insurance premiums are associated with states where drivers tend to have cleaner driving histories and fewer prior accidents.

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='abbrev', data=dataset)
plt.title('Total Car Crashes by State')
plt.xlabel('State')
plt.ylabel('Count')
plt.show()
```

The countplot showing the total car crashes by state reveals variations in the number of car crashes across different states. Some states have a notably higher count of car crashes, while others have a lower count. This plot provides a quick overview of the distribution of car crash counts by state, emphasizing the disparities in accident frequency among the states in the dataset.

```python
car_crashes_cile = np.ceil(dataset.loc[:, dataset.columns != 'abbrev'])
by_speeding = car_crashes_cile.groupby('speeding', as_index=False).sum()

sns.barplot(x='total', y='speeding', data=by_speeding);
```

Inference: The barplot suggests the relationship between the total number of car crashes and the speeding attribute. It appears to show how the total crashes are distributed across different levels or categories of speeding involvement.

# KDE PLOTS
KDE Plot described as Kernel Density Estimate is used for visualizing the Probability Density of a continuous variable. It depicts the probability density at different values in a continuous variable. We can also plot a single graph for multiple samples which helps in more efficient data visualization.

```python
sns.kdeplot(data=dataset,x="alcohol", fill=True)
```

The kernel density estimation (KDE) plot for the "alcohol" attribute displays the estimated probability density of alcohol involvement in car crashes. The filled KDE plot provides a smooth representation of the distribution of alcohol involvement, indicating that a significant portion of the data has relatively low alcohol involvement, with a peak in density around a specific value.

```python
# Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(x="abbrev", y="total", data=dataset)
plt.title("Violin Plot of Total Crashes by State")
plt.xlabel("State Abbreviation")
plt.ylabel("Total Crashes")
plt.show()
```

The violin plot visualizes the distribution of total crashes across different states using their abbreviations. It shows the range and distribution of crash counts in each state, with wider sections indicating a higher density of crash counts. The plot highlights variations in total crashes among states, with some states having a wider distribution and others having a narrower distribution of crash counts.

```python
# Pie Plot
state_counts = dataset['abbrev'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(state_counts, labels=state_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribution of Crashes by State")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

The pie plot illustrates the distribution of car crashes by state. It reveals the proportion of crashes in each state relative to the total, providing a visual representation of the state-wise distribution of car accidents in the dataset.

```python
# Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x="speeding", y="alcohol", data=dataset)
plt.title("Scatter Plot of Speeding vs. Alcohol Involvement in Crashes")
plt.xlabel("Speeding Involvement")
plt.ylabel("Alcohol Involvement")
plt.show()
```

The scatter plot depicts the relationship between speeding involvement and alcohol involvement in car crashes. It suggests that there is a positive correlation between speeding and alcohol involvement in crashes, as points tend to cluster towards the upper right corner, indicating that accidents with higher speeding involvement also tend to have higher alcohol involvement.

```python
# Create a bubble plot
plt.figure(figsize=(12, 8))
sns.scatterplot(x="total", y="alcohol", data=dataset, size="speeding", sizes=(10, 200), alpha=0.7)

# Customize labels and title
plt.xlabel("Total Crashes")
plt.ylabel("Alcohol-related Crashes")
plt.title("Bubble Plot of Car Crashes")

# Show the plot
plt.show()
```

The bubble plot visualizes the relationship between the total number of car crashes and alcohol-related crashes, with the size of each bubble representing the frequency of speeding-related crashes. It suggests that some states have a relatively high frequency of alcohol-related crashes, and in certain states with high total crashes, speeding-related crashes also appear to be prominent.

```python
# Create a swarm plot
plt.figure(figsize=(10, 6))
sns.swarmplot(x="total", y="alcohol", data=dataset)

# Customize labels and title
plt.xlabel("Total Crashes")
plt.ylabel("Alcohol-related Crashes")
plt.title("Swarm Plot of Car Crashes")

# Show the plot
plt.show()
```

The swarm plot provides a visual representation of the distribution of alcohol-related crashes across different total crash counts. It shows that while most states have relatively low total crashes and alcohol-related crashes, a few states exhibit higher alcohol-related crash rates, especially in cases of higher total crashes.

```python
# Create a strip plot
plt.figure(figsize=(10, 6))
sns.stripplot(x="total", y="alcohol", data=dataset , jitter=True, alpha=0.5)

# Customize labels and title
plt.xlabel("Total Crashes")
plt.ylabel("Alcohol-related Crashes")
plt.title("Strip Plot of Car Crashes")

# Show the plot
plt.show()
```

The strip plot indicates that as the total number of car crashes increases, there is a trend of higher occurrences of alcohol-related crashes, with some outliers indicating states with notably higher alcohol-related crash rates.
